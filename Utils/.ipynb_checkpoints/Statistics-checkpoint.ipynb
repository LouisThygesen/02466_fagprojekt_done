```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn; seaborn.set_style("darkgrid")
from tqdm.notebook import tqdm
import scipy.stats as st
import random
from itertools import combinations
import statsmodels.stats.multitest as bon 
import os

seed = 1
```

# Load and preprocess data:

```python
names = ['epoch', 'id', 'CTC_loss', 'WER', 'ref_WER_lens', 'hyp_WER_lens', 'CER', 'ref_CER_lens', 'hyp_CER_lens']
df_tc2 = pd.read_csv('tacotron2-cutvsdev-cleanY2021-m05-d07-H13-M43.csv', sep='\t', header=None, names=names)
df_base = pd.read_csv('test-clean-cutvsdev-cleanY2021-m05-d07-H13-M43.csv', sep='\t', header=None, names=names)
df_wavenet = pd.read_csv('wavenet-cutvsdev-cleanY2021-m05-d07-H13-M43.csv', sep='\t', header=None, names=names)

def set_error_rates(df):
    df['max_WER_lens'] = df['ref_WER_lens'] #df[['ref_WER_lens', 'hyp_WER_lens']].max(axis=1)
    df['WER_rate'] = df['WER']/df['max_WER_lens']
    df['max_CER_lens'] = df['ref_CER_lens'] #df[['ref_CER_lens', 'hyp_CER_lens']].max(axis=1)
    df['CER_rate'] = df['CER']/df['max_CER_lens']
    
set_error_rates(df_tc2)
set_error_rates(df_base)
set_error_rates(df_wavenet)
```

# Permuation test:

```python
epoch = 147
random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
np.random.seed(seed)

# Get data from last common epoch 
pd_wavenet = df_wavenet.loc[df_wavenet['epoch'] == epoch]
pd_tc2 = df_tc2.loc[df_tc2['epoch'] == epoch]
pd_base = df_base.loc[df_base['epoch'] == epch]

# Observe WER and CER (rates) 
pd_wavenet = pd_wavenet[['WER_rate','CER_rate']]
pd_tc2 = pd_tc2[['WER_rate','CER_rate']]
pd_base = pd_base[['WER_rate','CER_rate']]

# Collect CER and Wer for the 3 models 
list1 = [pd_wavenet['WER_rate'], pd_tc2['WER_rate'], pd_base['WER_rate']]
df_WER = pd.concat(list1,axis=1)

list2 = [pd_wavenet['CER_rate'], pd_tc2['CER_rate'], pd_base['CER_rate']]
df_CER = pd.concat(list2,axis=1)

def permutation_test(data1, data2):
    permutations = 6000
    
    # Calulate observed permutation statistic
    t_obs = data1.mean()-data2.mean()
    
    # Save t_0 distribution of permutation statistic samples 
    t_vals = np.zeros(permutations)
    
    # 1000 random permutations (with frozen transcript group)
    for k in range(permutations):
        perm = np.random.choice(a=[False, True], size=len(data1))
        
        group1 = perm*data1+np.logical_not(perm)*data2
        group2 = np.logical_not(perm)*data1+perm*data2
        
        s = group1.mean()-group2.mean()
        t_vals[k] = s
    
    # Calculate p-value
    logic_vector = (t_vals >= abs(t_obs)) + (t_vals <= -abs(t_obs))
    pval = np.mean(logic_vector)
        
    return t_obs, t_vals, pval
        
# Permutation tests on WER
t_obs1, t_vals1, p1 = permutation_test(df_WER.iloc[:,0],df_WER.iloc[:,1])
t_obs2, t_vals2, p2 = permutation_test(df_WER.iloc[:,1],df_WER.iloc[:,2])
t_obs3, t_vals3, p3 = permutation_test(df_WER.iloc[:,0],df_WER.iloc[:,2])

# Permutation tests on CER
t_obs4, t_vals4, p4 = permutation_test(df_CER.iloc[:,0],df_CER.iloc[:,1])
t_obs5, t_vals5, p5 = permutation_test(df_CER.iloc[:,1],df_CER.iloc[:,2])
t_obs6, t_vals6, p6 = permutation_test(df_CER.iloc[:,0],df_CER.iloc[:,2])

# Bonferroni correct p-values 
pvals1 = bon.multipletests([p1,p2,p3],alpha=0.05,method="bonferroni")[1]
pvals2 = bon.multipletests([p4,p5,p6],alpha=0.05,method="bonferroni")[1]
```

```python
for i in range(6):
    print("p-value {}: {}".format(i,pvals1[0]))
for i in range(6):
    print("p-value {}: {}".format(i,pvals2[0]))
```

```python
random.seed(1)
random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
np.random.seed(seed)

mapping = {"Taco": df_tc2.loc[ df_tc2.epoch == epoch, "WER_rate"],
           "Base": df_base.loc[ df_base.epoch == epoch, "WER_rate"],
           "Wavenet": df_wavenet.loc[ df_wavenet.epoch == epoch, "WER_rate"]}

possible_combinations = list(iter(combinations(["Taco", "Base", "Wavenet"], 2)))
triplets = []

for i, pair in enumerate(possible_combinations):
    print(pair)
    for model_a, model_b in [pair]:
        wer_a, wer_b = mapping[model_a], mapping[model_b]
        diff = wer_a - wer_b
        print(f"{model_a}-{model_b}")
        data = [ diff.sample(10000, replace=True).mean() for _ in range(1000) ]
        triplet = np.quantile(data, 0.025), np.mean(data), np.quantile(data, 0.975)
        triplets.append(triplet)
        print( [ round(t,5) for t in triplet] )  
```

# One-sample non-parametric bootstrapping:

Helper function:

```python
random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
np.random.seed(seed)

def extract_file_extension(file_name:str):
    """
    >>> extract_file_extensions("some_path/works_with_backslashes\\and_2x_extensions.tar.gz")
    '.tar.gz'
    """
    assert file_name.find(".") != -1, "No ´.´ found"
    suffixes = pathlib.Path(file_name).suffixes
    return ''.join(pathlib.Path(file_name).suffixes)

def save_plt_plot(save_path, fig=None, dpi=300):
    assert extract_file_extension(save_path) in [".png", ".jpg", ".pdf"]
    if fig is None:
        plt.savefig(save_path, dpi = dpi, bbox_inches = 'tight')
    else:
        fig.savefig(save_path, dpi = dpi, bbox_inches = 'tight')

class _ColorHEX:
    def __init__(self):
        self.blue = "#1f77b4"
        self.orange = "#ff7f0e"
        self.green = "#2ca02c"
        self.red = "#d62728"
        self.purple = "#9467bd" 
        self.brown = "#8c564b"
        self.pink = "#e377c2"
        self.grey =  "#7f7f7f"
colors_hex = _ColorHEX()
```

WER one-sample non-parametric bootstrapping confidence intervals for all epochs:

```python
random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
np.random.seed(seed)

# Bootstrap
df =  pd.DataFrame( np.ones((147*3, 4)), columns=["model_type", "mean", "low", "high"])
i = 0

for epoch in tqdm(range(1, 147)):
    mapping = {"tacotron": df_tc2.loc[ df_tc2.epoch == epoch, "WER_rate"],
               "baseline": df_base.loc[ df_base.epoch == epoch, "WER_rate"],
               "wavenet": df_wavenet.loc[ df_wavenet.epoch == epoch, "WER_rate"]}
    
    for model in ["tacotron", "baseline", "wavenet"]:
        wer = mapping[model]
        data = [ wer.sample(1000, replace=True).mean() for _ in range(1000) ]
        df.iloc[i] = model, np.mean(data), np.quantile(data, 0.025), np.quantile(data, 0.975)
        i += 1

        
# Plots
fig, ax = plt.subplots(figsize=(15, 5))
color_mapping = {"tacotron":colors_hex.blue, 
                 "baseline":colors_hex.orange, 
                 "wavenet":colors_hex.green}

for model in ["tacotron", "baseline", "wavenet"]:
    color = color_mapping[model]
    data = df[df["model_type"] == model].drop(columns=["model_type"]).to_numpy()

    xs = np.arange(len(data))

    mean, low, high = np.array(data)[:,0], np.array(data)[:,1], np.array(data)[:,2]

    ax.plot(mean, color=color, label=f'{model} - mean')
    plt.sca(ax) # <-- makes gca work, super wierd but gets the job done
    plt.gca().fill_between(xs, high, low,  color=color, alpha=0.5/3, label=f'{model} - 95% conf.')
    ax.legend()
    
plt.show()

plt.savefig("./bootstrap_wer.pdf", dpi = 180, bbox_inches = 'tight')
```

# t-test using CLT trick:

Method 1: One-sample t-test (meaning differences)

```python
list1 = np.array(list1)[:,:2700]
list2 = np.array(list2)[:,:2700]

def one_sample_t_test1(data,rows):
    diffs = data[rows[0],:] - data[rows[1],:]
    means = np.zeros(270)
    
    index = 0
    sum_10 = 0
    for k in range(270): 
        sum_10 += diffs[k]
        x = k%10
        
        if k%10==0:
            means[index] = sum_10/10
            index += 1
            sum_10 = 0
            
    stats = st.ttest_1samp(means, 0)
    
    return stats 
                  
# Test: WER
p1 = one_sample_t_test1(list1,[0,1])[1]
p2 = one_sample_t_test1(list1,[1,2])[1]
p3 = one_sample_t_test1(list1,[0,2])[1]

# Test: CER
p4 = one_sample_t_test1(list2,[0,1])[1]
p5 = one_sample_t_test1(list2,[1,2])[1]
p6 = one_sample_t_test1(list2,[0,2])[1]

# Bonferroni correct p-values 
pvals1 = bon.multipletests([p1,p2,p3],alpha=0.05,method="bonferroni")[1]
pvals2 = bon.multipletests([p4,p5,p6],alpha=0.05,method="bonferroni")[1]

for i in range(3):
    print("p-value {}: {}".format(i,pvals1[0]))
for i in range(3):
    print("p-value {}: {}".format(i,pvals2[0]))
```

```python
def one_sample_t_test2(data,rows):
    data1 = np.zeros(270)
    data2 = np.zeros(270)
    
    index = 0; sum1 = 0; sum2 = 0;
    
    for k in range(270): 
        sum1 += data[rows[0],k]
        sum2 += data[rows[1],k]
        
        x = k%10
        
        if k%10==0:
            data1[index] = sum1/10
            data2[index] = sum2/10
            index += 1; sum1 = 0; sum2 = 0;
    
    diffs = data1 - data2
    stats = st.ttest_1samp(diffs, 0)
    
    return stats 

# Test: WER
p1 = one_sample_t_test2(list1,[0,1])[1]
p2 = one_sample_t_test2(list1,[1,2])[1]
p3 = one_sample_t_test2(list1,[0,2])[1]

# Test: CER
p4 = one_sample_t_test2(list2,[0,1])[1]
p5 = one_sample_t_test2(list2,[1,2])[1]
p6 = one_sample_t_test2(list2,[0,2])[1]

# Bonferroni correct p-values 
pvals1 = bon.multipletests([p1,p2,p3],alpha=0.05,method="bonferroni")[1]
pvals2 = bon.multipletests([p4,p5,p6],alpha=0.05,method="bonferroni")[1]

for i in range(3):
    print("p-value {}: {}".format(i,pvals1[0]))
for i in range(3):
    print("p-value {}: {}".format(i,pvals2[0]))
```
